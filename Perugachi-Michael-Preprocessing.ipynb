#### 0. Cargar el Corpus

Vamos a trabajar con el corpus de Movie Reviews de IMDB

```python
import pandas as pd
```

```python
path = '/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv'
df = pd.read_csv(path, encoding = 'utf-8')
df
```

#### 1. Limpieza 

Limpiar los documentos de caracteres que no corresponden

```python
doc = df.iloc[0]['review']
doc
```

```python
doc.replace('', ' ')
doc = doc.replace('<br />', '').replace('>', '').replace('/', '')
doc
```

```python
import re
```

```python
def clean_text(doc):
    # Eliminar etiquetas HTML
    doc = re.sub(r"<.*?>", " ", doc)

    # Eliminar caracteres no alfabéticos
    doc = re.sub(r"[^a-zA-Z\s]", " ", doc)

    # Eliminar espacios múltiples
    doc = re.sub(r"\s+", " ", doc).strip()

    return doc
```

```python
df["clean"] = df["review"].apply(clean_text)
df["clean"]
```

```python
doc = df["clean"].iloc[0]
doc.split()
```

#### 2. Normalización

Convertir todos los tokens a minúsculas.

Elimina puntuación y símbolos no alfabéticos.

```python
df["normalized"] = df["clean"].str.lower()
df["normalized"]
```

```python
print(df["normalized"].iloc[0])
```

#### 3. Eliminación de Stopwords

Eliminar las palabras vacías (stopwords) usando una lista estándar de la librería _nltk_.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download("stopwords")
nltk.download("punkt")

stop_words = set(stopwords.words("english"))

## Funcion para Tokenizar y eliminar stopwords
def tokenize_and_remove_stopwords(doc):
    tokens = word_tokenize(doc)
    tokens = [t for t in tokens if t not in stop_words]
    return tokens

df["tokens"] = df["normalized"].apply(tokenize_and_remove_stopwords)
df["tokens"]
```

```python
print(df["tokens"].iloc[0])
```

#### 4. Stemming 

Reducir el espacio de palabras

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
def stem_tokens(tokens):
    return [stemmer.stem(t) for t in tokens]

df["stemmed"] = df["tokens"].apply(stem_tokens)
df["stemmed"]
```

```python
print(df["stemmed"].iloc[0])
```

#### 5. Verificar la diferencia

Comparar el tamaño del diccionario de términos del corpus antes y después de aplicar el preprocesamiento 

```python
# Vocabulario antes de stopwords y stemming

# Juntamos todos los documentos en un solo texto y lo dividimos en palabras individuales
vocab_original = set(" ".join(df["normalized"]).split()) 

# Vocabulario después

# Contruimos un teto con todos los tokens de todos los documentos 
vocab_final = set([t for doc in df["stemmed"] for t in doc])

print("Tamaño vocabulario original:", len(vocab_original))
print("Tamaño vocabulario procesado:", len(vocab_final))
```
